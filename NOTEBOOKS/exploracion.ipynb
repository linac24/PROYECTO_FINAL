#### Datos atípicos

Se realiza teniendo en cuenta el criterio de cuartiles

```python
Q1=Cuanti.quantile(0.25)
Q3=Cuanti.quantile(0.75)
IQR=Q3-Q1
Atipicos=((Cuanti<(Q1-1.5*IQR))|(Cuanti >(Q3+1.5*IQR)))
print(Cuanti[Atipicos.any(axis=1)])
```
