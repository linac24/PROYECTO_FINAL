A continuación se realiza una representación de las tres primeras componentes asociadas a las variables cuantitativas, sin embargo, no se logra una clara visualización , por ello se  realiza una observación de las dos primeras componentes en 2D.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Ajustar PCA a tus datos
pca = PCA(n_components=3)
X_pca = pca.fit_transform(CuantiScale)

# Crear el biplot en tres dimensiones
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar las observaciones en el espacio de las tres primeras componentes principales
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], alpha=0.6)


# Agregar las flechas que representan las variables originales
for i, feature in enumerate(Cuanti_new.columns):
    ax.quiver(0, 0, 0, pca.components_[0, i], pca.components_[1, i], pca.components_[2, i],
              color='red', arrow_length_ratio=0.1)
    ax.text(pca.components_[0, i]*1.15, pca.components_[1, i]*1.15, pca.components_[2, i]*1.15,
            feature, color='red', fontsize=12)

# Etiquetar los ejes
ax.set_xlabel('Primera componente principal')
ax.set_ylabel('Segunda componente principal')
ax.set_zlabel('Tercera componente principal')
ax.set_title('Biplot de PCA en tres dimensiones')

plt.show()
```

DBSCAN Clusterin

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Generar algunos datos de prueba
# Semilla

X = X_pca

# Instanciar y ajustar el modelo DBSCAN
# eps es la distancia máxima entre dos muestras para que se consideren en el mismo vecindario
# min_samples es el número de muestras en un vecindario para que un punto se considere como un punto central
dbscan = DBSCAN(eps=0.09, min_samples=3)
dbscan.fit(X)

# Etiquetas de los clusters
labels = dbscan.labels_

# Número de clusters en las etiquetas, ignorando el ruido si está presente
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Número estimado de clusters: %d' % n_clusters_)
print('Número estimado de puntos de ruido: %d' % n_noise_)

# Gráfico de los clusters y puntos de ruido
plt.scatter(X[:,0], X[:,1], c=labels,  marker='o', 
            edgecolor='k', s=50, alpha=0.5, )
plt.colorbar(label='Cluster')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Se observan dos grupos que resaltan, los amarillos y verdes, que posiblemente estan asociados a las personas con los niveles mas altos de obesidad. En donde la primer componente que inflyue es la relacionada con la edad y la segunda con los habitos salidables de alimentación, como el consumo de verduras, varias comidas principales, el consumod de agua.
